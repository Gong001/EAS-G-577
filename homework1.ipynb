##### Reading Temperature Data for Bloomington from the NOAA

```python
df = pd.read_csv('homework_data/Bloomington_temp_ave_indi_uni_station.csv')
```

##### Plotting the Probability Distributions of June, July, August, and Summer Average Temperatures for Bloomington

```python
fig, ax = plt.subplots(2, 2, figsize=(7, 5), dpi=300)

columns = ['Jun', 'Jul', 'Aug', 'mean']
titles = ['June', 'July', 'August', 'Summer Average']
labels = ['a)', 'b)', 'c)', 'd)']

for i, col in enumerate(columns):
    row, col_idx = divmod(i, 2)
    sns.histplot(df[col].dropna(), kde=True, color='#b22e01', ax=ax[row, col_idx], alpha=0.2, stat='probability')
    ax[row, col_idx].set_title(titles[i])
    ax[row, col_idx].set_xlabel('Temperature (Â°F)')
    ax[row, col_idx].set_xticks(np.arange(68, 83, 2))
    ax[row, col_idx].tick_params(axis='both', which='major', labelsize=10)
    ax[row, col_idx].set_yticks(np.arange(0, 0.4, 0.1))
    ax[row, col_idx].grid(linestyle='--', linewidth=0.2, alpha=0.5, color='k')
    ax[row, col_idx].text(0.03, 0.96, labels[i], transform=ax[row, col_idx].transAxes,
                          fontsize=12, verticalalignment='top')

plt.tight_layout()
plt.show()
```
