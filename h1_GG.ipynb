# Generalized Extreme Value (GEV) Distribution 

$$
F(x | \xi, \mu, \sigma) = \exp \left( - \left( 1 + \xi \frac{x - \mu}{\sigma} \right)^{-1/\xi} \right)
$$

Where:
- $\xi$ is the shape parameter,
- $\mu$ is the location parameter,
- $\sigma$ is the scale parameter.




```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import genextreme

# Define different parameter settings (shape, location, scale)
params = [
    (-0.5, 80, 2),  # shape, location, scale
    (0.0, 85, 1),
    (0.0, 85, 2),
    (0.0, 90, 2),
    (0.1, 80, 2)
]

# Generate x values (range of values where we want to evaluate the PDF)
x = np.linspace(70, 100, 1000)

# Set up the plot
plt.figure(figsize=(8, 4), dpi=300)

# Loop through each parameter setting and plot the corresponding GEV distribution
for param in params:
    shape, loc, scale = param
    # Generate the probability density function (PDF) for the GEV distribution
    cdf = genextreme.cdf(x, shape, loc, scale)
    plt.plot(x, cdf, label=f'shape={shape}, loc={loc}, scale={scale}')

# Add titles and labels
# plt.yticks([0.0, 0.1, 0.2, 0.3])
plt.title('GEV Distributions with Different Parameters')
plt.xlabel('Variable (Unit)')
plt.ylabel('Pumulative Distribution Function')
plt.legend()

plt.grid(ls='--')
plt.show()
```

- Changing the shape parameter affects the skewness of the CDF. For example, a shape parameter of `-0.5` results in a curve that rises steeply at lower values, indicating a left-skewed distribution where probabilities accumulate quickly at lower values.
- A location parameter of `80` places the curve more towards the left, while increasing it to `85` and then `90` shifts the curve to the right. This means as the location parameter increases, the distribution of variable values shifts rightward.
- The scale parameter affects the steepness of the CDF curve. A scale parameter of `1` results in a steeper curve, indicating more rapid accumulation of probability. Increasing the scale to `2` makes the curve flatter, showing a more gradual increase in probability over the range of variable values.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import genextreme

# Define different parameter settings (shape, location, scale)
params = [
    (-0.5, 80, 2),  # shape, location, scale
    (0.0, 85, 1),
    (0.0, 85, 2),
    (0.0, 90, 2),
    (0.1, 80, 2)
]

# Generate x values (range of values where we want to evaluate the PDF)
x = np.linspace(70, 100, 1000)

# Set up the plot
plt.figure(figsize=(8, 4), dpi=300)

# Loop through each parameter setting and plot the corresponding GEV distribution
for param in params:
    shape, loc, scale = param
    # Generate the probability density function (PDF) for the GEV distribution
    pdf = genextreme.pdf(x, shape, loc, scale)
    plt.plot(x, pdf, label=f'shape={shape}, loc={loc}, scale={scale}')

# Add titles and labels
# plt.yticks([0.0, 0.1, 0.2, 0.3])
plt.title('GEV Distributions with Different Parameters')
plt.xlabel('Variable (Unit)')
plt.ylabel('Probability Density Function')
plt.legend()

plt.grid(ls='--')
plt.show()
```

- A positive shape parameter (e.g., `0.1`) shifts the PDF curve to the right and makes it more concentrated towards higher values, indicating a left-skewed distribution.
- Increasing the location parameter to `90` shifts the peak of the PDF curve to the right, showing that the highest probability density is at higher variable values.
- A scale parameter of `1` results in a steeper PDF curve with a more pronounced peak and narrower tails.
